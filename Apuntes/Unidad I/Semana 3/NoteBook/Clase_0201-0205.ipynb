## Objetos e identificadores en Python
___
### Objetos
* Los objetos en python: son las formas en que Python representa los valores, todos los valores son objetos
* Enunciado de asignación: Establece un identificador o nombre y lo asocia con un objeto que está al lado derecho del signo igual. El tipo de dato se conoce a partir del literal del objeto. Compuesto de 3 elementos:
    1. Nombre o identificador asociado a un objeto (izq)
    2. Signo igual 
    3. Objeto, también puede ser función que al final se evalua como objeto (der)
* Ejemplo de asignación: 
```
temperatura = 35.4
```

### Identificadores
* Son sensibles a mayúsculas y minúsculas.
* Lo usual es definirlos usando cualquier combinación de letras, número y carácteres de guión bajo "_".
* No se puede iniciar el nombre de un identificador con un número 

#### Palabras reservadas (no se pueden usar para identificadores)

![palabras reservadas](https://drive.google.com/uc?export=view&id=1-nrNqi7-9WsfktgrRw-mw6_5RQs_EUwF)
    
### Tipado Dinámico
* Los identificadores pueden estar asociados a un tipo de objeto en un momento del tiempo y pueden esos identificadores asociarse a un tipo de objeto distinto en un momento posterior (No es una práctica conveniente)
* Es un lenguaje interactivo, se pueden escribir instrucciones en la terminal y obtener respuesta de Python
* Ejemplo: 

````python
#-*- coding: utf-8 -*-
"""
----------------------------------------------------------------------------------------------------
    
   
----------------------------------------------------------------------------------------------------
    Clase 0202 Ejemplo Tipado Dinámico 
----------------------------------------------------------------------------------------------------
"""

a = 'hola'
print(a)
a = 2
print(a)
a = True
if (a):
    print("Verdadero")
````

### Alias
* Es posible crear un alias asignando un segundo identificador a un mismo objeto
```
original = temperatura
```
* También luego es posible asignarlos a objetos diferentes (rompiendo el alias y creando un objeto nuevo)
```
temperatura += 5.0
```

### Clases predefinidas 

* Son clases que se pueden utilizar sin necesidad de importar ninguna libreria. 
* Las clases mutables: Los objetos creados en una clase pueden cambiar su valor
* Las clases inmutables: Los objetos creados en una clase no pueden cambiar su valor, por lo tanto debe ser sustituido por un objeto diferente 

| Tipo | Clase | Descripción | Inmutable |
| ------- |-------|-------------|-----------|
| Númerica | bool| valor boleano | x |
| Númerica | int | entero (magnitud arbitraria) | x |
| Númerica | float | Número punto flotante | x |
| Secuencias de elementos | list | Sequequencia de valores mutables | |
| Secuencias de elementos | tupla | Sequencia de valores inmutables | x |
| Secuencias de elementos | str | Cadena de caracteres | x |
| Conjuntos | set | Conjunto desordenado de distintos objetos | | 
| Conjuntos | frozenset | Forma inmutable del Conjunto de una clase | x |
| Conjuntos de pares clave:valor | dict | mapa asociativo (diccionario aka) | |

### Cases Númericas
#### Clase bool 

* Solamente tienen 2 posibles instancias: **True & False**
* Las clases tienen un constructor con el mismo nombre de la clase permiten crear objetos de tipo de la clase a partir de otros objetos. Ej Los números se evalúan a False si el número es es cero o a True si es distinto de cero
* El constructor bool() puede aceptar diferentes tipos de argumentos 
* Las secuencias (list, tuple, str) se evalúan a False si están vacías de lo contrario True

#### Clase int 

* No hay distintos tipos de entero, solo existe una clase que tiene magnitud arbitraria, es decir, el motor del lenguaje es el que se encargará de darle un almacenamiento suficiente para soportar el número entero el cual está limitado sólo por la memoria del computador
* La representación interna se selecciona autmomaticamente dependiendo del valor actual 
* El constructor de int() puede recibir numeros flotantes y los trunca Ej 1.6 = 2
* También puede recibir cadenas válidas y hace la conversión

#### Clase float

* Es la única clase que permite manejar números flotantes
* Tienen presición fija (no es arbitraria como en int) la limitación está dada por la implementación de Python y se debe validar el motor especifico en el que se esté trabajando
* el constructor float() acepta enteros y cadenas
* float(True) retorna 1.0, float(False) retorna 0.0

### Clases secuencias 

* Son clases que sirven como contenedores de otros objetos y a esos objetos les dan un orden predefinido
* Las clases secuencias más importantes en Py son: 
    1. list
    2. tuple
    3. str

#### Clase list
* Estructura referencial, las referencias (o apuntadores o direcciones de memoria de los objetos que pertenecen a la secuencia) de los objetos se almacenando de forma continua en la memoria, pero no los objetos como tal, sino los apuntadores o referencias de los objetos en si 
* Utilizan indices basados en cero 
* No existe tipo array de manera predefinida pero se puede importar
* Las listas pueden aumentar y disminuir su tamaño 

#### Clase tuple
* Secuencias inmutables, implica que no pueden cambiar de tamaño porque no pueden obtener un valor distinto al que se la al inicializarla. Esto permite operaciones más eficientes
* Igual que las listas, son basadas en referencias. 
* Utiliza parentesis 
* Tupla de un solo elemento igual usa coma
* La tupla puede almacenar una lista que sea mutable, porque la tupla lo que almacena es referencias de memoria, al igual que la lista, al modificar el tamaño de la lista su referencia de la tupla sigue siendo la misma 

#### Clase str 
* Secuencia inmutable de caracteres Unicode (permite usar caracteres de distintos alfabetos Ej Griego y Chino)
* Los carácteres pueden estar encerrados entre comillas dobles y simples
* El carácter contra-pleca ("\") sirve de carácter de escape, pera que forme parte de la cadena se debe escapar ("\\")
* Triple comilla doble o simple sirve para evitar escapar los saltos de línea en una cadena

````python
#-*- coding: utf-8 -*-

"""
----------------------------------------------------------------------------------------------------
    
 

----------------------------------------------------------------------------------------------------
    Ejemplos de las clases tipo secuncia
----------------------------------------------------------------------------------------------------
"""

# ------------------------------------------- Clase list -------------------------------------------
# Variable tipo lista - Método poco común 
# a = list(1, 'hola', True)

# Variable tipo lista - Método común 
a = [1, 'hola', True]

# Imprimir variable tipo lista
for i in range(len(a)):
    print(a[i])

# ------------------------------------------- Clase tuple -------------------------------------------
tupla = (1,2,3)

# Tupla solo con un elemento
tupla = (1,)

# ------------------------------------------- Clase str -------------------------------------------
cadena = "Don't Worry"
print(cadena)
cadena = "Don\'t Worry"
print(cadena)
cadena = "C:\\user\\downloads"
print(cadena)

# variable cadena que evita escapar los saltos de línea
cadena = """Este es un ejemplo de como evitar escapar los saltos de línea

Fin del ejemplo"""

# ------------------------------------------- En Terminal -------------------------------------------
# ------------------------------------------- Clase list -------------------------------------------
l = [1,2,3]
l[0] = 4
l
# ------------------------------------------- Clase tuple -------------------------------------------
tupla = (1,2,3)
tupla[0]
tupla = (1,2,l)
tupla[2][0] = 1
tupla
````

### Clases Conjuntos y Diccionarios
___

#### Clases de conjuntos
##### Clase set y frozenset
* Representan una colección de elementos únicos y sin un orden inherentes. 
* Están optimizados para consultar si un elemento pertenece o no al conjunto
* Sólo tipos inmutables pueden ser parte de un set o frozenset Ej No puede haber una lista dentro de un conjunto pero si una tupla 
* La diferencia entre set y frozen set es: 
    * Set es mutable, sus elementos deben ser inmutables, puede contener frozensets 
    * Frozenset es inmutable, no puede contener sets
* Las llaves "{}" vacias representan un diccionario vacio
* Es posible construir un conjunto a partir de una secuencia utilizando el constructor **set()**
* Si el constructor no recibe argumentos retorna un conjunto vacío
* Cuando se construye un conjunto con el constructor a partir de una lista, se eliminan los elementos repetidos
* Se puede crear una lista usando el constructor de list() a partir de un conjunto

#### Clases diccionario
##### Clase dict
* Representan una colección o un conjunto de pares clave:valor
* Las **claves o llaves** de un diccionario deben ser valores **inmutables**, Los valores no tienen este requisito. Las claves no se pueden repetir.
* Los elementos se delimitan usando llaves "{}" y los dos puntos ":"
* El constructor de un diccionario puede recibir una lista de pares para crearlo

````python
#-*- coding: utf-8 -*-

"""
----------------------------------------------------------------------------------------------------
    
   

----------------------------------------------------------------------------------------------------
    Ejemplos de las clases tipo conjunto
----------------------------------------------------------------------------------------------------
"""

# -------------------------------------------- Clase set -------------------------------------------
conjunto = {1, 2, 3, "Hola Mundo"}
print(conjunto)

# Se eliminan los elementos repetidos
conjunto = set([1, 1, 2, 2, 3, "Hola Mundo"])
print(conjunto)
 
# Representación de diccionario vacio
dict_vacio = {}
print(dict_vacio)
# -------------------------------------------- Clase dict ------------------------------------------
diccionario = {"a":1, "b":2}
print(diccionario)

# Lista de pares
diccionario = dict([("a", 1), ("b", 2)])
print(diccionario)
````
